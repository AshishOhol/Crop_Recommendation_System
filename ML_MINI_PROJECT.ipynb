```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Features and Target
X = Crop_recommendation_data[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]
y = Crop_recommendation_data['label'] 

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# Initialize scaler
scaler = StandardScaler()

# Fit ONLY on training data (Important!)
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data
X_test_scaled = scaler.transform(X_test)

print("Normalization Completed Successfully")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# Models
rf = RandomForestClassifier(n_estimators=200, random_state=42)
svm = SVC(probability=True, kernel='rbf', C=10, gamma='scale')
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# Train
rf.fit(X_train_scaled, y_train)
svm.fit(X_train_scaled, y_train)
xgb.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score

rf_acc = accuracy_score(y_test, rf.predict(X_test_scaled))
svm_acc = accuracy_score(y_test, svm.predict(X_test_scaled))
xgb_acc = accuracy_score(y_test, xgb.predict(X_test_scaled))

print("Random Forest Accuracy:", rf_acc)
print("SVM Accuracy:", svm_acc)
print("XGBoost Accuracy:", xgb_acc)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

knn_acc = accuracy_score(y_test, knn.predict(X_test_scaled))
print("KNN Accuracy:", knn_acc)


from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

nb = GaussianNB()

nb.fit(X_train_scaled, y_train)

y_pred_nb = nb.predict(X_test_scaled)

nb_acc = accuracy_score(y_test, y_pred_nb)

print("Naive Bayes Accuracy:", nb_acc)
```

```python
from sklearn.ensemble import VotingClassifier

ensemble = VotingClassifier(
    estimators=[
        ('rf', rf),
        ('svm', svm),
        ('xgb', xgb)
    ],
    voting='soft'   # important!
)

ensemble.fit(X_train_scaled, y_train)

ensemble_acc = accuracy_score(y_test, ensemble.predict(X_test_scaled))

print("Ensemble Accuracy:", ensemble_acc)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["Random Forest", "SVM", "XGBoost", "KNN", "Naive Bayes", "Ensemble"],
    "Accuracy": [rf_acc, svm_acc, xgb_acc, knn_acc, nb_acc, ensemble_acc]
})

print(results)
```

```python
from sklearn.preprocessing import label_binarize

n_classes = len(le.classes_)

y_test_bin = label_binarize(y_test, classes=range(n_classes))
```

```python
rf_probs = rf.predict_proba(X_test_scaled)
svm_probs = svm.predict_proba(X_test_scaled)
xgb_probs = xgb.predict_proba(X_test_scaled)
ens_probs = ensemble.predict_proba(X_test_scaled)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def plot_roc(y_test_bin, y_score, model_name):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Micro-average
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    plt.plot(fpr["micro"], tpr["micro"],
             label=f'{model_name} (AUC = {roc_auc["micro"]:.3f})')
```

```python
plt.figure(figsize=(8,6))

plot_roc(y_test_bin, rf_probs, "Random Forest")
plot_roc(y_test_bin, svm_probs, "SVM")
plot_roc(y_test_bin, xgb_probs, "XGBoost")
plot_roc(y_test_bin, ens_probs, "Ensemble")

plt.plot([0,1], [0,1], 'k--')  # diagonal line

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Multiclass - Micro Average)")
plt.legend()
plt.show()
```
